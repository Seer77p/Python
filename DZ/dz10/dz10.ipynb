```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
peng = sns.load_dataset("penguins")
```

```python
peng
```

```python
def f(row):
    if row['bill_length_mm'] < 35:
        val = 'low'
    elif row['bill_length_mm'] < 42:
        val = 'middle'
    else:
        val = 'high'
    return val

peng['beak_ength'] = peng.apply(f, axis = 1)
```
